```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.patches import Patch
```

```python
#dataframe import
df = pd.read_csv('conjunto_de_datos_ed2023.csv', sep=',')
df.head()
df.info()
```

```python
def graficar(df, x, y, title, xlabel, ylabel):
    """
    Función para graficar los datos de un DataFrame.
    
    Parámetros:
    df (DataFrame): DataFrame que contiene los datos a graficar.
    x (str): Nombre de la columna del eje x.
    y (str): Nombre de la columna del eje y.
    title (str): Título de la gráfica.
    xlabel (str): Etiqueta del eje x.
    ylabel (str): Etiqueta del eje y.
    """
    
    plt.figure(figsize=(10, 6))
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.axhline(0, color='black', lw=1)
    plt.axvline(0, color='black', lw=1)
    #Color
    sns.set_palette("Set2")
    #Estilo
    sns.set_style("whitegrid")
    #Tamano de fuente
    plt.rcParams.update({'font.size': 14})
    sns.scatterplot(data=df, x=x, y=y)
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.show()
```

```python
# Filtrar valores 999 en las columnas 'edad_div1' y 'edad_div2'
df = df[(df['edad_div1'] != 999) & (df['edad_div2'] != 999)]
df = df.dropna()
```

A. La relación que existe entre la edad y las personas que se divorcian en México.

```python
# Crear una nueva columna categórica para diferenciar x e y
df['categoria'] = ['x' if x > y else 'y' for x, y in zip(df['edad_div1'], df['edad_div2'])]

# Graficar con colores diferentes para x e y
sns.scatterplot(data=df, x='edad_div1', y='edad_div2', hue='categoria', alpha=0.5, s=17)

# Personalizar el gráfico
plt.xlabel('Persona 1')
plt.ylabel('Persona 2')
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
sns.set_style("whitegrid")
sns.set_palette("bright")
plt.title('Scatterplot de edad_div1 vs edad_div2 con colores diferentes')
plt.show()
```


B. La relación que existe entre los años que duró el matrimonio de las personas que se
divorciaron en México.

```python
# Calcular duración del matrimonio
df['duracion'] = df['anio_eje'] - df['anio_mat']
divorcios_duracion = df[(df['duracion'] >= 0) & (df['duracion'] <= 70)]
# Calcular estadísticas
media_duracion = divorcios_duracion['duracion'].mean()
y_max = divorcios_duracion['duracion'].value_counts().max()
# Gráfico de histogramas
plt.figure(figsize=(10, 6))
sns.histplot(divorcios_duracion['duracion'], bins=range(0, 71), color="#4c78a8", kde=False)
plt.axvline(media_duracion, color="red", linestyle="--", label=f"Duración media: {media_duracion:.1f} años")
plt.title("Distribución de la Duración de los Matrimonios hasta el Divorcio")
plt.xlabel("Años de duración del matrimonio")
plt.ylabel("Número de divorcios")
plt.legend()
plt.show()

```

D. La relación que existe entre los años que duró el matrimonio y las causas de divorcio en
México.

```python
# Filtrar registros con valores válidos
df = df[(df['anio_mat'] <= df['anio_eje']) & (df['anio_mat'] != 9999) & (df['anio_eje'] != 9999)]

# Calcular duración del matrimonio
df['duracion'] = df['anio_eje'] - df['anio_mat']

# Verificar si aún hay valores negativos
if (df['duracion'] < 0).any():
    print("Existen valores negativos en la duración. Revisa los datos.")
else:
    print("Todos los valores de duración son válidos.")

# Agrupar por causa y calcular la duración media
duracion_causa = df.groupby('descripcion')['duracion'].mean().reset_index()
duracion_causa = duracion_causa.sort_values(by='duracion', ascending=False)

# Graficar
plt.figure(figsize=(10, 6))
sns.barplot(data=duracion_causa, y='descripcion', x='duracion', palette="viridis")
plt.axvline(df['duracion'].mean(), color="red", linestyle="--", label=f"Duración media: {df['duracion'].mean():.1f} años")
plt.title("Duración media del matrimonio por causa de divorcio")
plt.xlabel("Duración media del matrimonio (años)")
plt.ylabel("Causa")
plt.legend()
plt.show()
```

E. La relación que existe entre la sexualidad de las parejas que se divorcian en México.

```python
# Calculamos los tipos de divorcios
heterosexuales = df[((df['sexo_div1'] == 1) & (df['sexo_div2'] == 2)) | 
                  ((df['sexo_div1'] == 2) & (df['sexo_div2'] == 1))]
homo_masculinos = df[(df['sexo_div1'] == 1) & (df['sexo_div2'] == 1)]
homo_femeninos = df[(df['sexo_div1'] == 2) & (df['sexo_div2'] == 2)]

# Creamos la figura
fig, ax = plt.subplots(1, 2, figsize=(15, 7))

# Gráfico de parejas heterosexuales (solo 1 barra)
ax[0].bar(['Parejas\heterosexuales'], 
         len(heterosexuales), 
         color='gray', 
         width=0.6)
ax[0].set_title('Divorcios Heterosexuales', fontsize=16, pad=20)
ax[0].set_ylabel('Cantidad de divorcios', fontsize=12)
ax[0].text(0, len(heterosexuales) + 50, 
          f"{len(heterosexuales):,}", 
          ha='center', 
          fontsize=12)

# Gráfico de parejas homosexuales
categorias = ['Parejas\ngay', 'Parejas\nlésbicas']
valores = [len(homo_masculinos), len(homo_femeninos)]
colores = ['blue', 'pink']

ax[1].bar(categorias, valores, color=colores, width=0.6)
ax[1].set_title('Divorcios Homosexuales', fontsize=16, pad=20)

# Añadimos valores encima de las barras
for i, valor in enumerate(valores):
    ax[1].text(i, valor + 5, f"{valor:,}", 
              ha='center', 
              fontsize=12)

# Configuraciones comunes
for axis in ax:
    axis.tick_params(axis='both', labelsize=12)
    axis.grid(axis='y', linestyle='--', alpha=0.4)
    
    # Quitamos bordes
    for spine in ['top', 'right']:
        axis.spines[spine].set_visible(False)

plt.suptitle('Distribución de Divorcios por Tipo de Pareja', 
             fontsize=20, 
             y=1.05)
plt.tight_layout()
plt.show()
```


H. La relación entre el estado de la república donde se efectuó el matrimonio y donde se
solicitó el divorcio.

```python
# Filtrar datos válidos para los campos 'ent_mat' (estado de matrimonio) y 'ent_regis' (estado de divorcio)
df = df[(df['ent_mat'].notna()) & (df['ent_regis'].notna())]

# Convertir los valores de 'ent_mat' y 'ent_regis' a enteros para facilitar el análisis
df['ent_mat'] = df['ent_mat'].astype(int)
df['ent_regis'] = df['ent_regis'].astype(int)

# Crear una tabla cruzada para contar las relaciones entre estados
tabla_estados = pd.crosstab(df['ent_mat'], df['ent_regis'])

# Crear un catálogo de estados para etiquetar los ejes
catalogo_entidades = {
    1: 'Aguascalientes', 2: 'Baja California', 3: 'Baja California Sur', 4: 'Campeche',
    5: 'Chiapas', 6: 'Chihuahua', 7: 'Coahuila', 8: 'Colima', 9: 'Durango', 10: 'Guanajuato',
    11: 'Guerrero', 12: 'Hidalgo', 13: 'Jalisco', 14: 'México', 15: 'Michoacán', 16: 'Morelos',
    17: 'Nayarit', 18: 'Nuevo León', 19: 'Oaxaca', 20: 'Puebla', 21: 'Querétaro', 22: 'Quintana Roo',
    23: 'San Luis Potosí', 24: 'Sinaloa', 25: 'Sonora', 26: 'Tabasco', 27: 'Tamaulipas',
    28: 'Tlaxcala', 29: 'Veracruz', 30: 'Yucatán', 31: 'Zacatecas', 32: 'CDMX'
}

# Renombrar índices y columnas de la tabla cruzada
tabla_estados.index = tabla_estados.index.map(catalogo_entidades)
tabla_estados.columns = tabla_estados.columns.map(catalogo_entidades)

import pandas as pd
import plotly.graph_objects as go

# Crear una tabla cruzada para contar las relaciones entre estados
tabla_estados = pd.crosstab(df['ent_mat'], df['ent_regis'])

# Crear listas para los nodos y flujos
source = []
target = []
value = []

for origen in tabla_estados.index:
    for destino in tabla_estados.columns:
        flujo = tabla_estados.loc[origen, destino]
        if flujo > 0:  # Solo incluir relaciones con flujo
            source.append(origen)
            target.append(destino)
            value.append(flujo)

# Crear etiquetas para los nodos
labels = [catalogo_entidades[i] for i in sorted(set(source + target))]

# Mapear índices de los nodos a las etiquetas
node_map = {estado: labels.index(catalogo_entidades[estado]) for estado in sorted(set(source + target))}
source_mapped = [node_map[s] for s in source]
target_mapped = [node_map[t] for t in target]

# Crear el diagrama de Sankey
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels
    ),
    link=dict(
        source=source_mapped,
        target=target_mapped,
        value=value
    )
)])

# Configurar el título y mostrar el gráfico
fig.update_layout(
    title_text="Relación entre el estado de matrimonio y el estado de divorcio",
    font_size=10
)
fig.show()
```
